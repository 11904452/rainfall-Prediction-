## Handle missing values

```python
df.info()
```

#### Impotation 
* Many attributes are present with null values or missing values, So i use imputation metjod to fill these null spaces

```python
# Missing value imputation using mean

df['MinTemp'].fillna(df['MinTemp'].mean(), inplace = True)
df['Rainfall'].fillna(df['Rainfall'].mean(), inplace = True)
df['Evaporation'].fillna(df['Evaporation'].mean(), inplace = True)
df['Sunshine'].fillna(df['Sunshine'].mean(), inplace = True)
df['Humidity9am'].fillna(df['Humidity9am'].mean(), inplace = True)
df['Humidity3pm'].fillna(df['Humidity3pm'].mean(), inplace = True)
df['Cloud9am'].fillna(df['Cloud9am'].mean(), inplace = True)
df['Cloud3pm'].fillna(df['Cloud3pm'].mean(), inplace = True)
df['Temp3pm'].fillna(df['Temp3pm'].mean(), inplace = True)
```

## Handle Outliers
Detecting outliers in a classification problem involves identifying data points that deviate significantly from the overall pattern of the dataset. 
* Use the seaborn or matplotlib library to create a box plot.
* Outliers are often considered as points beyond 1.5 times the interquartile range (IQR) from the first and third quartiles.

# Splitting data set into train and test

```python
# Splitting data into X and Y variable independent and dependent respectively
x = df.drop("RainTomorrow_Yes", axis=1)
x= x.drop("Date", axis=1)
y = df["RainTomorrow_Yes"]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
x_train.head()
```

# 3.Feature Scaling and Split train test data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train_s = scaler.fit_transform(x_train)
x_test_s = scaler.transform(x_test)
```

```python
x_train_s
```

```python
x_test_s
```

# 4. Model Selection

### 1. Logistic Regression

```python
# Importing Logistic Regression from Sklearn

from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train_s, y_train)
```

```python
# predicting test and train value over train dataset
y_train_lr = lr.predict(x_train_s)
y_pred_lr = lr.predict(x_test_s)
```

```python
print("Training dataset: ")
print("train Accuracy score: ",accuracy_score(y_train, y_train_lr))
print("train Precision score: ", precision_score(y_train, y_train_lr))
print("train Recall score: ", recall_score(y_train, y_train_lr))
print("Test dataset: ")
print("Test Accuracy score: ",accuracy_score(y_test, y_pred_lr))
print("Test Precision score: ", precision_score(y_test, y_pred_lr))
print("Test Recall score: ", recall_score(y_test, y_pred_lr))
```

```python
# Calulating and displaying confusion matrix

confusion_matrix(y_test, y_pred_lr)
```

```python
df_cm = pd.DataFrame(confusion_matrix(y_test, y_pred_lr), range(2), range(2))
sns.set(font_scale= 1.2) # for label size
sns.heatmap(df_cm, annot=True, annot_kws={"size": 14}) # font size

plt.show()
```

```python
# Plotting ROC curve

fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_lr)
auc = metrics.roc_auc_score(y_test, y_pred_lr) # Calculating and plotting auc value
plt.plot(fpr,tpr,label="LR, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

## 2. Decision Tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf_tree = DecisionTreeClassifier(max_depth = 3,
                                  random_state = 42)
clf_tree.fit(x_train_s, y_train)
```

```python
y_pred_dt = clf_tree.predict(x_test_s)
```

```python
print("Accuracy score for Decision tree:",accuracy_score(y_test, y_pred_dt))
print("Precision score for Decision tree:", precision_score(y_test, y_pred_dt))
print("Recall score for Decision tree:", recall_score(y_test, y_pred_dt))
```

```python
confusion_matrix(y_test, y_pred_dt)
```

```python
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_dt)
auc = metrics.roc_auc_score(y_test, y_pred_dt) # Calculating and plotting auc value
plt.plot(fpr,tpr,label="DT, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

## 3. Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rnd_clf = RandomForestClassifier(
    n_estimators=500,  
    min_samples_split=2, 
    min_samples_leaf=1,
    max_depth=15,
    n_jobs=-1, random_state = 42)
rnd_clf.fit(x_train_s, y_train)
```

```python
y_pred_rf = rnd_clf.predict(x_test_s)
```

```python
print("Accuracy score for Decision tree:",accuracy_score(y_test, y_pred_rf))
print("Precision score for Decision tree:", precision_score(y_test, y_pred_rf))
print("Recall score for Decision tree:", recall_score(y_test, y_pred_rf))
```

```python
confusion_matrix(y_test, y_pred_rf)
```

```python
# Plotting ROC curve

fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_rf)
auc = metrics.roc_auc_score(y_test, y_pred_rf) # Calculating and plotting auc value
plt.plot(fpr,tpr,label="LR, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

## hyperparameter tuning RAndom Forest 

```python
rnd_clf_2 = RandomForestClassifier()
```

```python
params_rf = {
    'n_estimators': [100, 200, 500],
    'max_depth': [None, 10, 15, 20],
    'min_samples_split': [2,4,6,8],
    'min_samples_leaf' : [1,3,5]
}
```

```python
# Performing grid search and storing in a model object
from sklearn.model_selection import GridSearchCV
grid_search_rf = GridSearchCV(rnd_clf_2, params_rf, cv = 5, n_jobs=-1)
```

```python
grid_search_rf.fit(x_train_s, y_train)
```

```python
# finding best param value using grid search attribute

grid_search_rf.best_params_
```

```python
rnd_best = grid_search_rf.best_estimator_
```

```python
y_pred_rf_best = rnd_best.predict(x_test_s)
```

```python
print("Accuracy score for Decision tree:",accuracy_score(y_test, y_pred_rf_best))
print("Precision score for Decision tree:", precision_score(y_test, y_pred_rf_best))
print("Recall score for Decision tree:", recall_score(y_test, y_pred_rf_best))
```

```python
confusion_matrix(y_test, y_pred_rf_best)
```

```python
# Plotting ROC curve

fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_rf_best)
auc = metrics.roc_auc_score(y_test, y_pred_rf_best) # Calculating and plotting auc value
plt.plot(fpr,tpr,label="LR, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
