# Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score,precision_score,recall_score, confusion_matrix
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.metrics import roc_curve, auc
```

```python
#Load data from csv file
df = pd.read_csv("sydney_rain prediction.csv", header=0)
```

# 2. Data Explore , visualization  and Preprocessing

## Handle Categorical data

```python
# Duumy variable creation

df = pd.get_dummies(df, columns = ["RainToday", "RainTomorrow"], drop_first = True)
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.astype({"RainToday_Yes":"int", "RainTomorrow_Yes":"int"})
```

## Correlation

```python
corr_matrix = df.corr()
corr_matrix
```

```python
corr_matrix["RainToday_Yes" ].sort_values(ascending=False)
```

* **Observation**

Min Temp and Temp9am both are independent attribute but both are highly correlated that can create multi collinearity in model so, i have to delete one of the attribute to avoid multi-collinearity. i delete "Temp9am" because it is less correlate with dependent variable as compare to Min Temp.

 Max Temp and Temp3pm both are independent attribute but both are highly correlated that can create multi collinearity in model so, i have to delete one of the attribute to avoid multi-collinearity. i delete "Temp3pm" because it is less correlate with dependent variable as compare to Max Temp.

 Max Temp and Temp9am both are independent attribute but both are highly correlated that can create multi collinearity in model so, i have to delete one of the attribute to avoid multi-collinearity. i delete "MAx Temp" because it is less correlate with dependent variable as compare to Temp9am.

Pressure9am and Pressure3pm both are independent attribute but both are highly correlated that can create multi collinearity in model so, i have to delete one of the attribute to avoid multi-collinearity. I decide to delete "Pressure3pm" because it is less correlate with dependent variable as compare to Pressure9am.


```python
# delete unnecessory data
del df['Temp9am']
del df['MaxTemp']
del df['Pressure3pm']
del df['Pressure9am']
del df['Location']
```

```python
# updated correlation matrix
corr_matrix = df.corr()
corr_matrix
```

```python
# Plotting heatmap of correlation matrix.

plt.figure(figsize = (15,15))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True)
```

## Handle missing values

```python
df.info()
```

```python
# Missing value imputation using mean

df['MinTemp'].fillna(df['MinTemp'].mean(), inplace = True)
df['Rainfall'].fillna(df['Rainfall'].mean(), inplace = True)
df['Evaporation'].fillna(df['Evaporation'].mean(), inplace = True)
df['Sunshine'].fillna(df['Sunshine'].mean(), inplace = True)
df['Humidity9am'].fillna(df['Humidity9am'].mean(), inplace = True)
df['Humidity3pm'].fillna(df['Humidity3pm'].mean(), inplace = True)
df['Cloud9am'].fillna(df['Cloud9am'].mean(), inplace = True)
df['Cloud3pm'].fillna(df['Cloud3pm'].mean(), inplace = True)
df['Temp3pm'].fillna(df['Temp3pm'].mean(), inplace = True)
```

# 3. Splitting data set into train and test

```python
# Splitting data into X and Y variable independent and dependent respectively
x = df.drop("RainTomorrow_Yes", axis=1)
x= x.drop("Date", axis=1)
y = df["RainTomorrow_Yes"]
```

```python
x.head()
```

```python
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
x_train.head()
```

# 4.Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_s = scaler.fit_transform(x_train)
x_test_s = scaler.transform(x_test)
```

```python
x_train_s
```

# 5. Train Model and  Model Evaluation

### 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train_s, y_train)
```

```python
# predicting test and train value over train dataset
y_train_lr = lr.predict(x_train_s)
y_pred_lr = lr.predict(x_test_s)
```

```python
print("Training dataset: ")
print("train Accuracy score: ",accuracy_score(y_train, y_train_lr))
print("train Precision score: ", precision_score(y_train, y_train_lr))
print("train Recall score: ", recall_score(y_train, y_train_lr))
print("Test dataset: ")
print("Test Accuracy score: ",accuracy_score(y_test, y_pred_lr))
print("Test Precision score: ", precision_score(y_test, y_pred_lr))
print("Test Recall score: ", recall_score(y_test, y_pred_lr))
```

```python
confusion_matrix(y_test, y_pred_lr)
```

```python

fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_lr)
auc = metrics.roc_auc_score(y_test, y_pred_lr) # Calculating and plotting auc value
plt.plot(fpr,tpr,label="Logistic, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

```python
confusion_matrix(y_test, y_pred_lr)
```

### 2.Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf_tree = DecisionTreeClassifier(max_depth = 3,
                                  min_samples_leaf=1,
                                  min_samples_split=2,
                                  random_state = 42)
clf_tree.fit(x_train_s, y_train)
```

```python
y_train_dt= clf_tree.predict(x_train_s)
y_pred_dt = clf_tree.predict(x_test_s)
```

```python
print("Training dataset: ")
print("train Accuracy score: ",accuracy_score(y_train, y_train_dt))
print("train Precision score: ", precision_score(y_train, y_train_dt))
print("train Recall score: ", recall_score(y_train, y_train_dt))
print("Test dataset: ")
print("Accuracy score: ",accuracy_score(y_test, y_pred_dt))
print("Precision score: ", precision_score(y_test, y_pred_dt))
print("Recall score: ", recall_score(y_test, y_pred_dt))
```

```python
confusion_matrix(y_test, y_pred_dt)
```

```python
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_dt)
auc = metrics.roc_auc_score(y_test, y_pred_dt) # Calculating and plotting auc value
plt.plot(fpr,tpr,label="DT, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

```python
from sklearn.tree import export_graphviz
from IPython.display import Image
import pydotplus

dot_data = export_graphviz(clf_tree,
                           out_file=None,
                           filled=True,
                           feature_names = x_train.columns,
                           rounded=True)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

## HyperParameter Tuning for decision Tree

```python
# Define hyperparameters for tuning
params_dt = {
    'max_depth': [3, 5, 10, 15],
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 2, 4, 6],
    'max_leaf_nodes': [None, 5, 10, 20]}
```

```python
# Perform Grid Search with 5-fold cross-validation
grid_search_t = GridSearchCV(estimator=clf_tree, param_grid=params_dt, cv=5)
grid_search_t.fit(x_train_s, y_train)
```

```python
grid_search_t.fit(x_train_s, y_train)
```

```python
grid_search_t.best_params_
```

```python
best_dt_model = grid_search_t.best_estimator_
```

```python
y_train_dt2 = best_dt_model.predict(x_train_s)
y_pred_dt2 = best_dt_model.predict(x_test_s)
```

```python
print("Training dataset: ")
print("train Accuracy score: ",accuracy_score(y_train, y_train_dt2))
print("train Precision score: ", precision_score(y_train, y_train_dt2))
print("train Recall score: ", recall_score(y_train, y_train_dt2))
print("Test dataset: ")
print("Accuracy score: ",accuracy_score(y_test, y_pred_dt2))
print("Precision score: ", precision_score(y_test, y_pred_dt2))
print("Recall score : ", recall_score(y_test, y_pred_dt2))
```

# 3. What were the available ML model options you had to perform this task?
**Logistic Regression:** While logistic regression is primarily used for binary classification, it can also be applied to multi-class classification tasks. It's a simple and interpretable model.

**knn:**
k-Nearest Neighbors (kNN) is a simple and effective algorithm used for both classification and regression tasks. It is a non-parametric, instance-based learning algorithm where the model makes predictions based on the k-nearest data points in the training set.

**Decision_tree:**
Decision Trees are versatile algorithms used for both classification and regression tasks. They work by recursively splitting the dataset into subsets based on the most significant attribute at each step. Import Decision tree from library scikit-learn.

**Bagging:**
Bagging, which stands for Bootstrap Aggregating, is an ensemble learning technique that involves training multiple models independently and combining their predictions. This helps improve the overall performance and robustness of the model. The most common form of bagging is Bootstrap Aggregating, where each model in the ensemble is trained on a random subset of the training data (with replacement).

**Random Forest:**
Random Forest, a popular machine learning algorithm, is an example of bagging. The main idea behind Random Forest is to train a collection of decision trees, each on a randomly sampled subset of the data (bagging), and then aggregate their predictions.

**Gradient Boosting:**
Gradient Boosting is another ensemble learning method that builds a sequence of weak learners (typically decision trees) to create a strong predictive model. Unlike Random Forest, Gradient Boosting builds trees sequentially, with each tree trying to correct the errors of the previous one. The main idea is to fit new models to the residuals (the differences between the observed and predicted values) of the existing model.

**Adaboost:**
AdaBoost (Adaptive Boosting) is an ensemble learning method that combines weak learners (typically decision trees) to create a strong predictive model. AdaBoost assigns different weights to each observation in the dataset, and it focuses on the examples that the previous weak models misclassified. The final model is a weighted sum of weak models, with more weight given to models that perform well on the training data.

**XGBoost:**
XGBoost (eXtreme Gradient Boosting) is an efficient and scalable implementation of gradient boosting designed for speed and performance. It is widely used for both classification and regression tasks and is known for its accuracy and efficiency. XGBoost is an extension of gradient boosting that introduces additional regularization terms and parallel processing to improve model performance.
